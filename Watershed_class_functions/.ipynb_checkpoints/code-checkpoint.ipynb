```python
#Load the watershed
wat = wmf.SimuBasin(rute='watershed.nc')
```

## Plots of the watershed 

### Plot only the watershed divisory

```python
#plot of the watershed divisory
wat.plot_basin()
```

### Plot watershed divisory along with the network

```python
#Save the network to a shp file 
wat.Save_Net2Map('network.shp')

#Plot with it 
wat.plot_basin(shape_path='network.shp', shape_color='b', shape_width=2)
```

### Plot watershed with network and a raster variable

```python
#Get the stream and hills Horton order 
wat.GetGeo_StreamOrder()
#Get the HAND values
wat.GetGeo_HAND_and_rDUNE()
```

**Note:** usually the results from the Watershed functions store the obtained variables as properties of the watershed with a name starting as wat.Cell... . In this case the computed variables are: wat.Cellhorton_Hill and wat.CellHAND. 

```python
res = wat.plot_basin(wat.CellHorton_Hill,)
```

wat.plot_basin(var) plots by default the given variable using the **viridis** color scheme (default in matplotlib). It also sets by default the values and the localization of the colorbar. The function has several arguments to customize its behavior, we show some of them in the following example: 

```python
res = wat.plot_basin(wat.CellHorton_Hill,
    cmap=pl.get_cmap('YlGnBu',4),
    cbar_loc=[0.3,0.1,0.5,0.02], 
    cbar_ticks=[1,2,3,4],cbar_title='Hills Horton order',
    shape_path='network.shp', shape_color='b', shape_width=2)
```

In this case, the variable **res** contains the result of the function. If we took a look to the variable we will observe how it has the axis object, the colorbar object, and the coordinates. the user can use those for a further customization of the watershed plot.

```python
#Axis
res[0]
```

```python
#colorbar
res[1]
```
