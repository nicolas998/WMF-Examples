# Import Packages 

```python
%matplotlib inline 
from wmf import wmf
import numpy as np 
import pylab as pl
import pandas as pd 
import glob 
```

# Interpolated fields

```python
#Read the watershed structure
cu = wmf.SimuBasin(rute='watershed.nc')
```

```python
#Read the coordinates and ids of the IDEAM stations near the Miel Watershed.
coord, stations = wmf.read_map_points('estaciones_ideam.shp', ListAtr=['COD_CATALO'])
stations = stations['COD_CATALO'].tolist()

#Function to obtain the id of the files in the folder.
def extract_ids(path):
    return path.split('/')[-1].split('_')[-1].split('.')[0]

# Get the list of stations in the folder.
ListStations = glob.glob('rain_records/P_*.msgpack')
# Get the id of the stations in the folder
ids_inFile = list(map(extract_ids, ListStations))

#Get the shared stations
shared_stations, idx_sta, idx_File = np.intersect1d(np.array(stations), np.array(ids_inFile), return_indices=True)
```

```python
#Read the information from pandas, and record the coordinates and ids of the stations that already has information.
Good_ids = []
Good_coord = []
Data = []
for idx_f, idx_s in zip(idx_File, idx_sta):
    try:
        d = pd.read_msgpack(ListStations[idx_f])
        if d['2016'].values.size == 366:
            Data.append(d['2016'].values)
            Good_ids.append(stations[idx_s])
            Good_coord.append(coord.T[idx_s])
            dates = d['2016'].index
    except:
        pass
Data = np.array(Data)
Rain = pd.DataFrame(Data.T, index=dates)
Good_coord = np.array(Good_coord)
```

```python
#Interpolate rainfall with the IDEAM stations.
RainMean = cu.rain_interpolate_idw(Good_coord.T, Rain, 'Miel_rain_v1.bin', p = 3)
```

Time series of the interpolated rainfall filed

```python
# Read the obtained table of the interpolated rainfall, it contains the record id of the binary and the mean rainfall.
Rain = wmf.read_rain_struct('Miel_rain_v1.hdr')
fig = pl.figure(figsize=(15,5), dpi = 250)
ax = fig.add_subplot(111)
ax.tick_params(labelsize = 16)
Rain[' Lluvia'].plot.line(fig = fig, ax = ax, lw = 3)
ax2 = ax.twinx()
ax2.tick_params(labelsize = 16)
ax2.set_xticklabels([])
Rain['acum'] = Rain[' Lluvia'].cumsum()
Rain['acum'].plot.line(fig = fig, ax = ax2, c = 'r',lw = 3)
ax.set_ylabel('Rain [$mm$]', size = 20)
pl.savefig('Annual_rainfall.png', bbox_inches = 'tight')
```

Reads the binary with the rainfall records and plot the total rainfall for the period.

```python
#Example of how to read the records of each record at the binary file.
#set an empty vector with the number of cells of the watershed
RainTotal = np.zeros(cu.ncells)
#Iterate trough the records to obtain the values
for rec in Rain[' Record'].values:
    if rec > 1:
        rain, status = wmf.models.read_int_basin('Miel_rain_v1.bin', rec, cu.ncells)
        RainTotal += rain.astype(float) / 1000 #Divide by 1000 since the records are multiplied by 1000 when stored 
```

```python
cu.Save_Net2Map('network.shp')
```

```python
res = cu.plot_basin(RainTotal, shape_path='network.shp', shape_color='b', shape_width=1,
        cmap=pl.get_cmap('YlGnBu',5),
        cbar_loc=[0.3,0.1,0.5,0.02], 
        cbar_ticks=[1600, 2400,3200,3200+800,3200+800*2],cbar_title='Rainfall [$mm$]')
```

# Radar derived fields

```python
#Read a new watershed 
cu = wmf.SimuBasin(rute='turkeyBasin150.nc')

#Reads the tif files with the coordinates of the Stage 4 rainfall 
stX, p, e = wmf.read_map_raster('s4_x1.tif')
stX = cu.Transform_Map2Basin(stX, p)
stX = stX.astype(int)

stY, p, e = wmf.read_map_raster('s4_y.tif')
stY = cu.Transform_Map2Basin(stY, p)
stY = stY.astype(int)

stage4 = stX*1000+stY
stage4 = stage4.astype(int)
```

```python
#Reads the data 
data = pd.read_msgpack('st4_rainfall.msg')
#Gets the dates
dates = pd.to_datetime(np.unique(data.index))
#Creates a continuos date range 
dates2 = pd.date_range(dates[0], dates[-1], freq='1H')
```

```python
#Creates the rainfall for a random slice of period (to reduce the weight)
for date in dates2[4000:5000]:
    rain = np.zeros(stage4.shape)
    try:
        if type(data.loc[date]['id']) == pd.core.series.Series:
            val, idx1, idx2 = np.intersect1d(data.loc[date]['id'].values, np.unique(stage4), return_indices=True)
            if val.size > 0:
                for idx, value in zip(val, data.loc[date]['val'].values[idx1]):
                    if value > 1000: value = 0.0
                    p = np.where(stage4 == idx)[0]
                    rain[p] = value
    except:
        pass
    cu.rain_radar2basin_from_array(rain,'rain_2016_turkey', date)
cu.rain_radar2basin_from_array(status='close', ruta_out='rain_2016_turkey.hdr')
```

```python
#Read a slice of rainfall and plot it 
rain, status = wmf.models.read_int_basin('rain_2016_turkey.bin', 23, cu.ncells)
```

```python
res = cu.plot_basin(rain / 1000.,
    cmap=pl.get_cmap('YlGnBu',5),cbar_title='Rain [$mm$]')
```
